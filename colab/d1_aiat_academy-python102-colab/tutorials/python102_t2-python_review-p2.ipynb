# **T2.3: การหา Factorial โดยการใช้ลูป และเงื่อนไข** 

**n! = n \* (n-1) \* (n-2) \* ... \* 1**

Note:
+ รับค่าตัวเลขที่ต้องการหา Factorial จากคีย์บอร์ด
+ หาค่า Factorial โดยการใช้ลูป
+ คืนค่าจากฟังก์ชั่น factorial พร้อมแสดงผลออกทางหน้าจอ


```
def factorial(number):
  result = 1
  # looping towrds the number variable reversely
  for num in reversed(range(number+1)): #(5, 4, 3, 2, 1, 0)
    if num != 0:
      result = result * num
  return result
  

n = int(input("Enter a number: "))
n_fact = factorial(n)
print("Factorial({}) is {}.".format(n, n_fact))
```

```
def factorial(number):
  result = 1
  for num in range(1, number+1): #n = 5: (1, 2, 3, 4, 5)
      result = result * num
  return result
  

n = int(input("Enter a number: "))

n_fact = factorial(n)
print("Factorial({}) is {}.".format(n, n_fact))
print("Factorial(", n, ") is",n_fact,".")
```

```
months = ["Janurary", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]

sorted_months_asc = sorted(months)
print("Sorted months (Ascending):", sorted_months_asc)

sorted_months_dsc = sorted(months, reverse=True)
print("Sorted months (Descending):", sorted_months_dsc)


numbers = [5, 4, 0, 1, 7, 5]
sorted_numbers_asc = sorted(numbers)
print("Sorted numbers (Ascending):", sorted_numbers_asc)

sorted_numbers_dsc = sorted(numbers, reverse=True)
print("Sorted numbers (Descending):", sorted_numbers_dsc)
```

**List Appending and Removing**


การเพิ่มข้อมูลและลบข้อมูลของลิสต์

```
lst = [5, 4, 0, 1, 7, 5]
lst2 = ['Hi', 'My', 'Name', 'is', 'Ben']


#ต่อท้าย: append(value)
lst.append(-1)
print(lst)
#[5, 4, 0, 1, 7, 5, -1]

#นำข้อมูลออกนอก List: pop(index) ถ้าไม่ได้ระบุ index จะนำตัวสุดท้ายออกมากจาก list
lst.pop()
print(lst)
#[5, 4, 0, 1, 7, 5]

lst.append("str")
print(lst)
#[5, 4, 0, 1, 7, 5, "str"]


#ลบตัวที่ระบุ: remove(value)
lst.remove("str")
print(lst)
#[5, 4, 0, 1, 7, 5]

lst.remove(lst[0])
print(lst)
#[4, 0, 1, 7, 5]

#แทรกค่าเข้าไปในสมาชิกของ list: insert(index, value)
lst.insert(3, 'Hello')
print(lst)
#[4, 0, 1,'Hello', 7, 5]

lst.append(lst2)
print(lst) #list ซ้อน list จะเรียกว่า 'Nested list'

# ขณะนี้ lst คือ [4, 0, 1, 'Hello', 7, 5, ['Hi', 'My', 'Name', 'is', 'Ben']]
#ต้องการเข้าถึงค่า 'Ben'

name = lst[6][4]
print(name)

lst = [1, 2, 3, 4]
lst2 = [9, 8, 7, 6, 5]

#การนำ element มาเชื่อมกันได้โดยใช้ extend, เครื่องหมาย(+)
lst.extend(lst2)
print(lst)

newlst = lst + lst2
print(newlst)

duplst = lst*10
print(duplst)
```

# T2.5: **การใช้ลูปกับลิสต์**

การสร้างลิสต์ของรายชื่อนักเรียนและใช้ลูปในการแสดงผลข้อมูลในลิสต์

```
def get_listlength(lst):
  return len(lst)

if __name__ == '__main__':
  students = []

  # วนรับรายชื่อนักเรียนจากคีย์ยอร์ดจนกว่าจะพิมพ์ "Q" หรือ "q"
  while(True):
    user_input = input("Enter a student name: ")

    if user_input.lower() == 'q':
      break

    students.append(user_input)

  # วนลูปเพิ่อแสดงรายชื่อนักเรียนทั้งหมดจาก list ออกทางหน้าจอ
  print("List of students:")
  for student in students:
    print(student)

  print("Number of students:", get_listlength(students))
  
```

**Sort dictionary**

การจัดเรียงลำดับในดิกชันนารี

```
mydict = {'carl':40, 'alan':2, 'bob':1, 'danny':3}

print(mydict.keys())
print(mydict.values())

#'Sort by keys
for key in sorted(mydict.keys()):
    print ("%s: %s" % (key, mydict[key]))
```

```
#Sort by keys
for key, value in sorted(mydict.items(), key=lambda item: item[0]):
    print ("%s: %s" % (key, value))
print("---------------------")
#Sort by items
for key, value in sorted(mydict.items(), key=lambda item: item[1]):
    print ("%s: %s" % (key, value))
```
