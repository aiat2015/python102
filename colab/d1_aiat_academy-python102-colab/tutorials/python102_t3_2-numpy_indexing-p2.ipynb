# NumPy Indexing and Selection

การเข้าถึงอาเรย์ของ NumPy

```python
import numpy as np
```

```python
#Creating sample array (สร้าง Array)
array = np.arange(0,11)
print(array)
```

## Bracket Indexing and Selection
The simplest way to pick one or some elements of an array looks very similar to python lists:
(วิธีการเข้าถึงค่าใน Element ของ NumPy Arrays จะคล้ายกับ List ของ Python)

```python
#Get a value at an index
array[5]
```

```python
#Get values in a range (Slicing)
print(array[1:6])
print(array[0:6])
```

## Broadcasting

Numpy arrays differ from a normal Python list because of their ability to broadcast:

การกระจายค่าเข้าไปใน NumPy Arrays

```python
#Setting a value with index range (Broadcasting)
array[0:5]=100
print(array)
```

```python
# Reset array
array = np.arange(0,11)
print(array)
```

```python
#Important notes on Slices
slice_of_array = array[0:6]
print(slice_of_array)
```

```python
#Change Slice

slice_of_array[:]=99
print(slice_of_array)
```

Now note the changes also occur in our original array!

การเปลี่ยนค่าของ Elements ใน บางส่วนของ Array ใหม่ (**slice_of_array**) จะเปลี่ยนค่า Original อง Array (**array**) นั้นด้วย

```python
print(array)
```

Data is not copied, it's a view of the original array! This avoids memory problems!

ซึ่งหมายความว่า Python ไม่ได้ Copy ค่าจาก Original Array มาใส่ที่ ตัวแปรใหม่ (ตัวแปรใหม่ (**slice_of_array**) ยังมี Address ในการเก็บข้อมูลเดียวกันกับ Original Array  (**array**))

```python
#To get a copy, need to be explicit
array_copy = array.copy()
print(array_copy)
```

## Indexing a 2D array (matrices)

การเข้าถึง อาเรย์ 2 มิติ

The general format is **arr_2d[row][col]** or **arr_2d[row,col]**. 

```python
array_2d = np.array(([5,10,15],[20,25,30],[35,40,45]))
print(array_2d)
```

```python
#Indexing row 
# เข้าถึงแถวที่ 1 Note: Index เริ่มจาก 0
print(array_2d[1])
```

```python
# Format is array_2d[row][col] or array_2d[row,col]
# Getting individual element value
print(array_2d[1][0])
```

```python
# Getting individual element value
print(array_2d[1,0])
```

```python
# 2D array slicing
#Shape (2,2) from top right corner
print(array_2d[:2,1:])
```

```python
#Shape bottom row
print(array_2d[2])
```

```python
#Shape bottom row
print(array_2d[2,:])
```

### Fancy Indexing

Fancy indexing allows you to select entire rows or columns out of order,to show this, let's quickly build out a numpy array:

```python
#Set up matrix
array2d = np.zeros((10,10))
```

```python
#Length of array
array_length = array2d.shape[1]
print(array_length)
```

```python
#Set up array

for i in range(array_length):
    array2d[i] = i
    
print(array2d)
```

Fancy indexing allows the following

```python
array2d[[2,4,6,8]]
```

```python
#Allows in any order
array2d[[6,4,2,7]]
```
